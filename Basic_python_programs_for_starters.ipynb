```
# question 13
# Integer variables
num_int1 = 10
num_int2 = 5

# Float variables
num_float1 = 15.5
num_float2 = 3.0

# Complex variables
num_complex1 = 2 + 3j
num_complex2 = 1 - 2j


# Integer Operations
sum_int = num_int1 + num_int2
print(f"Integer Sum: {sum_int}, Type: {type(sum_int)}")

diff_int = num_int1 - num_int2
print(f"Integer Difference: {diff_int}, Type: {type(diff_int)}")

prod_int = num_int1 * num_int2
print(f"Integer Product: {prod_int}, Type: {type(prod_int)}")

div_int = num_int1 / num_int2
print(f"Integer Division: {div_int}, Type: {type(div_int)}") # Result is always float


# Float Operations
sum_float = num_float1 + num_float2
print(f"Float Sum: {sum_float}, Type: {type(sum_float)}")

prod_float = num_float1 * num_float2
print(f"Float Product: {prod_float}, Type: {type(prod_float)}")


# Mixed Type Operations (Integer + Float)
mixed_sum = num_int1 + num_float1
print(f"Integer + Float Sum: {mixed_sum}, Type: {type(mixed_sum)}") # Result is float

mixed_prod = num_int1 * num_float1
print(f"Integer * Float Product: {mixed_prod}, Type: {type(mixed_prod)}") # Result is float


# Complex Operations
sum_complex = num_complex1 + num_complex2
print(f"Complex Sum: {sum_complex}, Type: {type(sum_complex)}")

prod_complex = num_complex1 * num_complex2
print(f"Complex Product: {prod_complex}, Type: {type(prod_complex)}")
```

```
# question 17
print("========== Demonstrating Immutability (Strings) ==========")
my_string = "Hello"
print(f"Original String: {my_string}, ID: {id(my_string)}")

# Attempting to modify a character in a string (will raise an error if uncommented)
# my_string[0] = 'J' # This line would cause a TypeError

# To 'modify' a string, a new string object is created
my_string = my_string + " World"
print(f"Modified String: {my_string}, ID: {id(my_string)}")


print("\n========== Demonstrating Mutability (Lists) ==========")
my_list = [1, 2, 3]
print(f"Original List: {my_list}, ID: {id(my_list)}")

# Modifying an element in the list
my_list[0] = 10
print(f"List after element modification: {my_list}, ID: {id(my_list)}")

# Adding an element to the list
my_list.append(4)
print(f"List after appending element: {my_list}, ID: {id(my_list)}")
```

```
# question 18
# Integer variable
int_var = 10

# Float variable
float_var = 3.14

# String variable
str_var = "Python Programming"

# Boolean variable
bool_var = True

# List variable
list_var = [1, 2, 3, "four"]

# Tuple variable
tuple_var = (10, 20, 30)

# Dictionary variable
dict_var = {"name": "Alice", "age": 30}

# Set variable
set_var = {1, 2, 3, 3, 4}

# NoneType variable
none_var = None

variables = [
    ("int_var", int_var, int),
    ("float_var", float_var, float),
    ("str_var", str_var, str),
    ("bool_var", bool_var, bool),
    ("list_var", list_var, list),
    ("tuple_var", tuple_var, tuple),
    ("dict_var", dict_var, dict),
    ("set_var", set_var, set),
    ("none_var", none_var, type(None))
]

print("--- Data Type Checks using type() and isinstance() ---")
for name, value, expected_type in variables:
    print(f"Variable: {name:<10}, Value: {str(value):<20}, Actual Type: {type(value).__name__:<10}, Is {expected_type.__name__}? {isinstance(value, expected_type)}")

print(f"\nExample: Is int_var a float? {isinstance(int_var, float)}")
print(f"Example: Is str_var a list? {isinstance(str_var, list)}")
```

```
# question 19
print("========== Demonstrating id() function ==========")

# Create two variables with the same value
value_a = 100
value_b = 100

print(f"Initial state:")
print(f"value_a = {value_a}, ID: {id(value_a)}")
print(f"value_b = {value_b}, ID: {id(value_b)}")


print("\n========== After Reassignment of value_a ==========")

# Reassign value_a to a different value
value_a = 200

print(f"value_a = {value_a}, ID: {id(value_a)}")
print(f"value_b = {value_b}, ID: {id(value_b)}")
```
