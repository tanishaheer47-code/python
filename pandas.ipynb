```python
#IS NULL
import pandas as pd
cities = {"London": 8615246,
"Berlin": 3562166,
"Madrid": 3165235,
"Rome": 2874038,
"Paris": 2273305,
"Vienna": 1805681,
"Bucharest": 1803425,
"Hamburg": 1760433,
"Budapest": 1754000,
"Warsaw": 1740119,
"Barcelona": 1602386,
"Munich": 1493900,
"Milan": 1350680}
city_series = pd.Series(cities)
print(city_series)
my_cities = ["London", "Paris", "Zurich", "Berlin",

"Stuttgart", "Hamburg"]
my_city_series = pd.Series(cities,

index=my_cities)

print("\nis null?\n " ,my_city_series.isnull())

# NOT NULL
print("\nNOT NULL\n : ", my_city_series.notnull())
```

```python
# to drop null values
print(my_city_series.dropna())
```

```python
# to fill blank spaces
missing_cities = {"Stuttgart":597939, "Zurich":378884}
my_city_series.fillna(missing_cities)
```

```python
# NA as integer value
my_city_series = my_city_series.fillna(0).astype(int)
print(my_city_series)
```

```python
#dataframes from list of list
import pandas as pd

# Define a list of lists (data)
data_lists_new = [
    [10, 'X', False],
    [20, 'Y', True],
    [30, 'Z', False],
    [40, 'a' , True],
    [50, 'b' , True],
    [60, 'c' , True],
    [70, 'd' , True],
    [80, 'e' , True],
    [90, 'f' , True],
    [100, 'g' , True]

]

columns_new = ['id', 'category', 'status']

df_from_lists_new = pd.DataFrame(data_lists_new, columns=columns_new)

print(df_from_lists_new)
```

```python
# Identify all columns that contain 'paid' (case-insensitive) to handle potential leading/trailing spaces
columns_to_drop = [col for col in df_from_lists_new.columns if 'paid' in col.lower()]

# Drop all identified 'paid' columns
if columns_to_drop:
    df_from_lists_new.drop(columns=columns_to_drop, inplace=True)
    print("DataFrame after dropping 'paid' columns:")
    print(df_from_lists_new)
    print("\nUpdated column names:")
    print(df_from_lists_new.columns)
else:
    print("No 'paid' columns found to drop.")
```
